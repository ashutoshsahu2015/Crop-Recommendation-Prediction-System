### Feature Engineering

```python
## Separting feature and target label
features = dataset[['NITROGEN', 'PHOSPHORUS', 'POTASSIUM', 'TEMPERATURE', 'HUMIDITY', 'PH','RAINFALL']]
target= dataset['CROP']
```

```python
## Splitting into training and testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,target,test_size=0.2,random_state=2)
```

### Model Training

```python
## Intializing the array for accuracy and model
acc = []
model = []
```

```python
# Decision Tree

from sklearn.metrics import classification_report
from sklearn import metrics

from sklearn.tree import DecisionTreeClassifier

DecisionTree = DecisionTreeClassifier(criterion="entropy",random_state=2,max_depth=5)

DecisionTree.fit(X_train,y_train)

train_values = DecisionTree.predict(X_train)
predicted_values = DecisionTree.predict(X_test)


x = metrics.accuracy_score(y_test, predicted_values)
y = metrics.accuracy_score(y_train, train_values)

acc.append(x)
model.append('Decision Tree')
print('Train Accuracy is: ',y*100)
print("DecisionTrees's Accuracy is: ", x*100)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(DecisionTree, features, target,cv=5)
score
```

```python
## Guassian Naive Bayes
from sklearn.naive_bayes import GaussianNB

NaiveBayes = GaussianNB()

NaiveBayes.fit(X_train,y_train)

train_values = NaiveBayes.predict(X_train)
predicted_values = NaiveBayes.predict(X_test)

x = metrics.accuracy_score(y_test, predicted_values)
y= metrics.accuracy_score(y_train,train_values)

acc.append(x)
model.append('Naive Bayes')

print("Training Accuracy is: ",y*100)
print("Naive Bayes's Accuracy is: ", x*100)
```

```python
score = cross_val_score(NaiveBayes,features,target,cv=5)
score
```

```python
## Support Vector Machine (SVM)

from sklearn.svm import SVC

SVM = SVC(gamma='auto')

SVM.fit(X_train,y_train)

train_values = SVM.predict(X_train)
predicted_values = SVM.predict(X_test)

x = metrics.accuracy_score(y_test, predicted_values)
y = metrics.accuracy_score(y_train,train_values)

acc.append(x)
model.append('SVM')

print("Training Accuracy is: ",y*100)
print("SVM's Accuracy is: ", x*100)
```

```python
score = cross_val_score(SVM,features,target,cv=5)
score
```

```python
## Logistic Regression
from sklearn.linear_model import LogisticRegression

LogReg = LogisticRegression(random_state=2)

LogReg.fit(X_train,y_train)

train_values = LogReg.predict(X_train)
predicted_values = LogReg.predict(X_test)


x = metrics.accuracy_score(y_test, predicted_values)
y = metrics.accuracy_score(y_train, train_values)

acc.append(x)
model.append('Logistic Regression')

print("Training Accuracy is: ",y*100)
print("Logistic Regression's Accuracy is: ", x*100)
```

```python
# Cross validation score (Logistic Regression)
score = cross_val_score(LogReg,features,target,cv=5)
score
```

```python
## Random Forest
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(n_estimators=20, random_state=0)
RF.fit(X_train,y_train)

train_values = RF.predict(X_train)
predicted_values = RF.predict(X_test)


x = metrics.accuracy_score(y_test, predicted_values)
y= metrics.accuracy_score(y_train,train_values)

acc.append(x)
model.append('Random Forest')

print("Training Accuracy is: ",y*100)
print("RF's Accuracy is: ", x*100)
```

```python
score = cross_val_score(RF,features,target,cv=5)
score
```

```python
## K-Nearest Neighbor

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,y_train)

train_values = knn.predict(X_train)
predicted_values = knn.predict(X_test)

x = metrics.accuracy_score(y_test,predicted_values)
y = metrics.accuracy_score(y_train,train_values)

acc.append(x)
model.append("K-Nearest Neighbor")

print('Training Accuracy is: ',y*100)
print("K-Nearest Neighbor Accuracy is: ",x*100)
```

```python
score = cross_val_score(knn,features,target,cv=5)
score
```
